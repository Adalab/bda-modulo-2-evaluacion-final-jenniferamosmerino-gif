## Ejercicio 1. Creación de una bases de datos de películas extraídas de un API

```python
import requests
import numpy as np
import pandas as pd
import mysql.connector
from mysql.connector import Error
```

### Fase 1: Extracción de datos de películas

```python
url = "https://beta.adalab.es/resources/apis/pelis/pelis.json"
```

```python
def llamar_api(endpoint):
    datos = requests.get(endpoint)
    print(f"llamar al api con respuesta : {datos.status_code}")

    if datos.status_code != 200:
        print(f"llamada falló por el siguiente motivo {datos.reason}")

    else:
        return datos.json()
```

```python
datos_p = llamar_api(url)
```

```python
datos_p
```

```python
Datos_películas = []
```

```python
Datos_películas = []

for película in datos_p:
    adultos = ()
    if película["adultos"] == True:
       adultos = "Sí"
    if película["adultos"] == False:
       adultos = "No"

    subtitulos = ",".join(película["subtitulos"])
 

    nueva_película = {"Título": película["titulo"].capitalize().strip(),
                      "Año": película["año"],
                      "Duración": película["duracion"],
                      "Género": película["genero"].capitalize().strip(),
                      "Contenide_adultos": adultos,
                      "Subtítulos": subtitulos
    }
   
    Datos_películas.append (nueva_película)
```

```python
Datos_películas
```

```python
DfDatos_películas= pd.DataFrame(Datos_películas)
DfDatos_películas
```

### Fase 3: Inserción de los Datos en la Base de Datos

```python
query = """
INSERT INTO películas (
    Título,Año, Duración, Género, Contenido_adultos, Subtítulos)
 VALUES (%s, %s, %s, %s, %s, %s)
"""
```

### Fase 4: Obtener información a partir de los datos.

```python
#### 1.¿Cuántas películas tienen una duración superior a 120 minutos?
query = '''SELECT Título AS Título_película, Duración AS Duración_película
        FROM películas
        WHERE Duración > 120
        ORDER BY Duración;'''

df = pd.read_sql(query, connection)
df
```

```python
#### 2.¿Cuántas películas incluyen subtítulos en español?
query = f'''SELECT COUNT(*) AS Total_películas_esp
        FROM películas
        WHERE Subtítulos LIKE '%ES%';'''

df = pd.read_sql(query, connection)
df
```

```python
#### 3.¿Cuántas películas tienen contenido adulto?
query = '''SELECT COUNT(*) AS Total_adulto
        FROM películas
        WHERE Contenido_adultos = 'Sí';'''

df = pd.read_sql(query, connection)
df
```

```python
#### 4.¿Cuál es la película más antigua registrada en la base de datos?
query = '''SELECT Título, Año
        FROM películas
        ORDER BY Año
        LIMIT 1;'''

df = pd.read_sql(query, connection)
df
```

```python
#### 5.Muestra el promedio de duración de las películas agrupado por género.
query = '''SELECT ROUND(AVG(Duración)) AS Duración_media, Género
        FROM películas
        GROUP BY Género
        ORDER BY Duración_media DESC;'''

df = pd.read_sql(query, connection)
df
```

```python
#### 6.¿Cuántas películas por año se han registrado en la base de datos? Ordena de mayor a menor (NUMERO DE PELÍCULAS)
query = '''SELECT COUNT(Título) AS Num_Películas, Año
        FROM películas
        GROUP BY Año
        ORDER BY Num_Películas DESC;'''

df = pd.read_sql(query, connection)
df
```

```python
#### 7.¿Cuál es el año con más películas en la base de datos?
query = '''SELECT COUNT(Título) AS Num_Películas, Año
        FROM películas
        GROUP BY Año
        ORDER BY Num_Películas DESC
        LIMIT 1;'''

df = pd.read_sql(query, connection)
df
```

```python
#### 8.Obtén un listado de todos los géneros y cuántas películas corresponden a cada uno.
query = '''SELECT Género, COUNT(Título) AS Num_Películas
        FROM películas
        GROUP BY Género
        ORDER BY Género;'''

df = pd.read_sql(query, connection)
df
```

```python
#### 9.Muestra todas las películas cuyo título contenga la palabra "Godfather" (puedes usar cualquier palabra).
query = '''SELECT Título
        FROM películas
        WHERE Título LIKE '%The%';'''

df = pd.read_sql(query, connection)
df
```
